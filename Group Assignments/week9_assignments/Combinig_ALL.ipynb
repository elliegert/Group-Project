## Combining All data to one dataframe for final

```python
#step 1: import libraries

# for data wrangling
import pandas as pd

# for spatial
import geopandas as gpd

# for plotting
import matplotlib.pyplot as plt

# for basemaps
import contextily as ctx

# for interactive plots
import plotly.express as px
```

```python
#Speed limit geojson
gdf_speed = gpd.read_file('Data/speeds_location_buffer_1.geojson')

#Street type geojson
gdf_streets = gpd.read_file('Data/street_type.geojson')

#Street light geojson
gdf_lights = gpd.read_file('Data/lights_buffer_timecode.geojson')
```

```python
gdf_speed.head(2)
```

```python
gdf_streets.head(2)
```

```python
gdf_lights.head(2)
```

```python
#merging to one dataframe
```

```python
#merge new dataframe to original dataframe to get geometry column back
gdf_master=gdf_speed.merge(gdf_streets,on='ID')
gdf_master.head(3)
```

```python
gdf_master=gdf_master.merge(gdf_lights,on='ID')
```

```python
gdf_master.head(2)
```

```python
#write to file to save this 
gdf_master.to_csv("speeds_streets_lights_locations_buffer.csv")
```
